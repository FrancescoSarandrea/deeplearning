Transform the Growth Rate into a binary variable

```python
def isGrowing(number):
    if number>1:
        return 1
    else:
        return 0
```

```python
dfSmaller['Growth Rate']
growingDfSmaller=dfSmaller['Growth Rate'].map(isGrowing)
```

```python
growingDfSmaller
dfSmaller['Growth Rate']=growingDfSmaller.astype('int')
```
